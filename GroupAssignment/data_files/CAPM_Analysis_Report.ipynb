# CAPM Analysis Report
This report provides a detailed summary and visualization of CAPM regression results across firms.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CAPM results
capm_results = pd.read_csv("capm_results_set1.csv")
capm_results.head()
```

```python
# Summary statistics
capm_results.describe()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(capm_results['alpha'], bins=30, kde=True)
plt.title("Distribution of CAPM Alpha")
plt.xlabel("Alpha")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(capm_results['beta'], bins=30, kde=True)
plt.title("Distribution of CAPM Beta")
plt.xlabel("Beta")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(capm_results['r_squared'], bins=30, kde=True)
plt.title("Distribution of CAPM R")
plt.xlabel("R")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=capm_results, x='beta', y='alpha')
plt.title("Alpha vs Beta (CAPM)")
plt.xlabel("Beta")
plt.ylabel("Alpha")
plt.grid(True)
plt.tight_layout()
plt.show()
```
